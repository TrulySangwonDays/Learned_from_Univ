# 유튜브 인기급상승 제목&조회수
---

```python
from selenium import webdriver

# run the one of the code lines below according to your OS type
# driver = webdriver.Chrome('./chromedriver') # mac
driver = webdriver.Chrome() # windows

driver.implicitly_wait(3) # waiting 3 seconds
```

```python
# access to target url
driver.get('https://www.youtube.com/feed/trending?bp=6gQJRkVleHBsb3Jl') 
```

---

```python
driver.find_element("css selector", "#video-title > yt-formatted-string").text
```

```python
driver.find_elements("css selector", "#video-title > yt-formatted-string")

#video-title > yt-formatted-string
```

```python
elements = driver.find_elements("css selector", '#video-title')

# news attribute lists
title_list = []

for x in elements:
    # title
    title = x.find_element("css selector", 'yt-formatted-string').text
    # print
    print(title)
    # append news item values to the attribute lists
    title_list.append(title)
```

```python
driver.find_element("css selector", "#content").text


#video-title > yt-formatted-string
#video-title
#video-title > yt-formatted-string
#video-title
#description-text
#content-text
#content
```

---

```python
soup = bs(driver.page_source, 'lxml')
```

```python
title = soup.select('a#video-title')
# 영상 제목만 전체 조회
for i in title:
    print(i.text.strip())
```

```python
soup.select('span.style-scope.ytd-video-meta-block')
```

```python
view = soup.select('div#metadata-line > span:nth-child(3)')
```

```python
title_list = []
view_list = []

for i in range(len(title)):
    title_list.append(title[i].text.strip())
    view_list.append(view[i].text.strip())

print(title_list)
print(view_list)
```

```python
info = {'제목':title_list, '조회수':view_list}
pd.DataFrame(info).head(50)
```
