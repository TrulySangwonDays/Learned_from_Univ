## Problem 1. Summary statistics

Using the Titanic dataset (located in `./data/titianic.csv`), identify the following:

1. mean and median of `Age`?
2. mean and median of `Passenger Fare`?
3. the number of missing values in the `Age` and `Passenger Fare` columns?

- **HINT**: to check the number of missing values of a column, use `.isna().sum()` method. (`isna()` checks whether a given value is missing or not. `sum()` sums up given values.)
- If you’re having trouble, search Google for similar Q&As. (e.g., StackOverflow)
- Note: round to the nearest integer.

```python
df1 = pd.read_csv('./titanic.csv')
df1.head()
```

```python
print("1. mean and median of Age")
print("mean of Age :", round(df1['Age'].mean()))
print("median of Age :", round(df1['Age'].median()))
print("")
print("2. mean and median of Passenger Fare")
print("mean of Passenger Fare :", round(df1['Passenger Fare'].mean()))
print("median of Passenger Fare :", round(df1['Passenger Fare'].median()))
print("")
print("3. the number of missing values in the Age and Passenger Fare columns")
print("the number of Mv in the Age :", df1['Age'].isna().sum())
print("the number of Mv in the Passenger Fare :", df1['Passenger Fare'].isna().sum())
```

## Problem 2. Correlation between Sleep Time and GPA

The `exercise1_sleep_gpa.csv` (located in `./data/`) contains sleep time (`Sleep`) and GPA (`GPA`) of 20 individuals.

1. Find the correlation coefficient between the two variables (round to 2 digits).
2. Draw a scatter plot and describe the relationship between the two variables. Do they have a linear relationship? Is it week or strong?

```python
df2 = pd.read_csv('./sleep_gpa.csv')
df2.head()
```

```python
#1. Correlation coefficient between the two variables

df2_corr = df2.corr()
round(df2_corr, 2)
```

```python
#2. The relationship between the two variables

#. import package
import seaborn as sns

sns.scatterplot(x = df2['Sleep'],
               y = df2['GPA']).set(title = "Scatter between the two variables")

print("Sleep와 GPA 두 변수 간의 상관계수는 0.69로 뚜렷한 양적 선형관계라고 판단할 가능성이 크다.")
```

## Problem 3. Find the two most similar stocks.

`exercise1_stock.csv` (located in `./data/`) contains the daily stock prices of 4 companies for 30 consecutive days.

Find the two most similar two stocks showing similar up and down patterns and explain the why.

```python
df = pd.read_csv('./stock.csv')
df
```

```python
print("우선 네 종목 간 페어플롯을 통해 상관관계를 시각화 해본다.")
sns.pairplot(df, height = 2)
```
