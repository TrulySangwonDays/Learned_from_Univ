## Problem 1. Join and GroupBy

Suppose you want to concatenate two datasets, `exercise1_boxA.csv` and `exercise1_boxB.csv`.

```python
df_a = pd.read_csv('boxA.csv')
df_a
```

```python
df_b = pd.read_csv('boxB.csv')
df_b
```

However, you realized that the `price` column is missing in `exercise1_boxB` dataframe!

The price info of all products can be found in `exercise1_price.csv`.

```python
df_p = pd.read_csv('price.csv')
df_p
```

#### 1. Add the `price` column to `exercise1_boxB` using the `exercise1_price` table.


```python
df_b = pd.merge(df_b, df_p, how = 'left', left_on = 'ProductCode', right_on = 'ProductCode')
df_b
```

#### 2. Concatenate the two tables, `exercise1_boxA.csv` and `exercise1_boxB.csv` into one table and drop duplicate rows. 

Hint: use `.drop_duplicates()` method. After that, there will be 15 rows remained.

```python
df = pd.concat([df_a, df_b], axis = 0).drop_duplicates()
df
```

#### 4. Find the mean and median of `Weight_gram` and `Price` grouped by `Membership`.

Hint: use `groupby()` method.

Note: Round to the nearest integer.

```python
# mean
round(df.groupby('Membership').mean(), 1)
```

```python
# median
round(df.groupby('Membership').median(), 1)
```
