```python
from selenium import webdriver
driver = webdriver.Chrome ('chromedriver.exe')
driver.implicitly_wait (3) # waiting 3 seconds
```

```python
# access to target url
driver.get('https://sports.news.naver.com/wfootball/news/index?isphoto=N')
```

```python
driver.find_element("css selector",
                    "#_newsList > ul > li:nth-child(1) > div > a > span").text
```

```python
title_objects = driver.find_elements("css selector",
                    "#_newsList > ul > li > div > a > span")
```

```python
# title
for obj in title_objects :
    print(obj.text)
```

```python
#_newsList > ul > li > div > a
```

```python
url_objects = driver.find_elements("css selector",
                    "#_newsList > ul > li > div > a")
```

```python
# url 가져오기
for obj in url_objects :
    print(obj.get_attribute('href'))
```

```python
li_list = driver.find_elements("css selector",
                    "#_newsList > ul > li")
li_list
```

```python
title_list = []
url_list = []
datetime_list = []

for li in li_list :
    title = li.find_element("css selector",
                                "div > a > span").text
    url = li.find_element("css selector",
                                "div > a").get_attribute('href')
    datetime = li.find_element("css selector",
                                "div > div > span.time").text
    print(title, url, datetime)
    title_list.append(title)
    url_list.append(url)
    datetime_list.append(datetime)
```

```python
text_list = []

for url in url_list :
    #access
    driver.get(url)
    #get text
    text = driver.find_element("css selector",
                        "#newsEndContents").text
    text_list.append(text)
    print(text)
    print()
    
```

```python
len(text_list)
```

```python
import pandas as pd

df = pd.DataFrame({'title' : title_list,
                  'text' : text_list,
                  'datetime' : datetime_list,
                  'url' : url_list
                  })
```
